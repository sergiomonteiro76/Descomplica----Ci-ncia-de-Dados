**Importação dos Pacotes**

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

**Dados de Entrada**

```python
x=[10,15,16,18]
y=[100,180,200,300]
```

# **Regressão Linear**

```python
modelo_linear = np.poly1d(np.polyfit(x, y, 1))
linha = np.linspace(1, 22, 100)
plt.scatter(x, y,c="red")
plt.plot(linha, modelo_linear(linha))
plt.show()
```

# **Regressao Polinomial**

```python
modelo_polinomial = np.poly1d(np.polyfit(x, y, 3))
linha = np.linspace(1, 22, 100)
plt.scatter(x, y,c="red")
plt.plot(linha, modelo_polinomial(linha))
plt.show()
```
